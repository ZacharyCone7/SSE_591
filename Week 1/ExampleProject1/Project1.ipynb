```python
global result, num1, num2
result = 0; num1 = 0; num2 = 0; i = 0

print("Select an option")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
print("5. Exponent")
print("6. Inverse")
print("99. Main Menu")

def input_options():
    global result, num1, num2
    if result == 0:
        num1 = int(input("What is the first input?"))
        num2 = int(input("What is the second input?"))
    else:
        num1 = int(input("What is the second input?"))

def add():
    global result, num1, num2
    if result == 0:
        input_options()
        print(f"The sum of the two inputs is = {num1 + num2}")
        result = (num1 + num2)
    else:
        input_options()
        result += num1
        print(f"The sum of the two inputs is = {result}")

def subtract():
    global result, num1, num2
    if result == 0:
        input_options()
        print(f"The difference of the two inputs is = {num1 - num2}")
        result = (num1 - num2)
    else:
        input_options()
        result -= num1
        print(f"The difference of the two inputs is = {result}")

def multiply():
    global result, num1, num2
    if result == 0:
        input_options()
        print(f"The product of the two inputs is = {num1 * num2}")
        result = (num1 * num2)
    else:
        input_options()
        result *= num1
        print(f"The product of the two inputs is = {result}")

def divide():
    global result, num1, num2
    if result == 0:
        input_options()
        print(f"The quotient of the two inputs is = {num1 / num2}")
        result = (num1 / num2)
    else:
        input_options()
        result /= num1
        print(f"The quotient of the two inputs is = {result}")

def exponent():
    global result, num1, num2
    if result == 0:
        input_options()
        print(f"The exponential of the two inputs is = {num1 ** num2}")
        result = (num1 ** num2)
    else:
        input_options()
        result **= num1
        print(f"The exponential of the two inputs is = {result}")

def inverse():
    global result
    result = -result
    print(f"The inverse of the result is = {result}")
               
user_select = input("Select an option:")
while True:
    if user_select == "1":
        add()
    elif user_select == "2":
        subtract()
    elif user_select == "3":
        multiply()
    elif user_select == "4":
        divide()
    elif user_select == "5":
        exponent()
    elif user_select == "6":
        inverse()
    elif user_select == "99":
        break
    else:
        print("Invalid input")
        i = 1
    if i == 1:
        user_select = input("Re-enter input") 
        i = 0
    else:
        user_select = input("Select an option:")
```
