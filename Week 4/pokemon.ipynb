### Pokemon Data Analysis and Visualization

#### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

poke_df = pd.read_csv(r"C:\Users\Zachary.Cone.SERN\Desktop\SSE 591-Intro to Python & Data Science\Week 4\SSE591_Week4\pokemon.csv")

display(poke_df)
```

```python
# Filter PokÃ©mon with Type 1 as Water
water_pokemon = poke_df[poke_df['Type 1'] == 'Water']

# Count occurrences of each Type 2
type2_counts = water_pokemon['Type 2'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(type2_counts, labels=type2_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Type 2 Distribution for Water Type PokÃ©mon')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

#### Read in datasets

#### Line Plot

```python
poke_df.Speed.plot(kind = 'line', color = 'g', label = 'Speed', linewidth=1,
                   alpha=0.5, grid=True, linestyle=':')
poke_df.Defense.plot(color='r', label='Defense', linewidth=1, alpha=0.5,
                     grid=True, linestyle='-.')
plt.legend(loc='upper right')
plt.xlabel('Pokemon Index')
plt.ylabel('Stat')
plt.title('Line Plot');
```

```python
plt.plot(poke_df['HP'])
plt.xlabel('Pokemon Index')
plt.ylabel('HP')
plt.title('Distribution of Pokemon HP');
```

```python
# Filter PokÃ©mon with Type 1 or Type 2 as Steel
steel_pokemon = poke_df[(poke_df['Type 1'] == 'Steel') | (poke_df['Type 2'] == 'Steel')]

# Count the number of Steel-type PokÃ©mon in each generation
steel_counts = steel_pokemon['Generation'].value_counts()

# Plotting the histogram
plt.bar(steel_counts.index, steel_counts.values, color='steelblue')
plt.xlabel('Generation')
plt.ylabel('Number of Steel-type PokÃ©mon')
plt.title('Number of Steel-type PokÃ©mon in Each Generation')
plt.xticks(steel_counts.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Box plot comparing the distributions of Pokemon Attack Stats across different Legendary statuses

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='Legendary', y='Attack', data=poke_df)
plt.xlabel('Legendary')
plt.ylabel('Attack')
plt.title('Attack Distribution by Legendary Status');
```

```python
# Sum up the values from 'HP' to 'Speed' for each PokÃ©mon
poke_df['Total'] = poke_df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].sum(axis=1)

# Print the DataFrame with the new 'Total' column
print(poke_df[['Name', 'Total']])

plt.figure(figsize=(12, 5))
sns.violinplot(x='Generation', y='Total', data=poke_df)
plt.xlabel('Generation')
plt.ylabel('Total')
plt.title('Distribution of Stats Across Generations')
plt.show()
```

```python
sns.pairplot(poke_df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def']])
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```
