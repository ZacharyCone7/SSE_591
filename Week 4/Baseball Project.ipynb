```python
# Extract relevant columns
years = BB_df['Year']
home_runs = BB_df['HR']

# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(years, home_runs, marker='o', linestyle='-')

# Add title and labels
plt.title('Barry Bonds Career Home Runs')
plt.xlabel('Year')
plt.ylabel('Home Runs')

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Extract relevant columns
BH_years = BH_df['Year']
BH_hits = BH_df['H']

MT_years = MT_df['Year']
MT_hits = MT_df['H']


# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(BH_years, BH_hits, '-o', label='Bryce Harper')
plt.plot(MT_years, MT_hits, '-s', label='Mike Trout')

# Add title and labels
plt.title('Harper & Trout Career Hit Comparisons')
plt.xlabel('Year')
plt.ylabel('Hits')

# Show the plot
plt.grid(True)
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)

plt.show()
```
